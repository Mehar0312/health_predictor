```python
cols_names={"bp":"blood_pressure",
          "sg":"specific_gravity",
          "al":"albumin",
          "su":"sugar",
          "rbc":"red_blood_cells",
          "pc":"pus_cell",
          "pcc":"pus_cell_clumps",
          "ba":"bacteria",
          "bgr":"blood_glucose_random",
          "bu":"blood_urea",
          "sc":"serum_creatinine",
          "sod":"sodium",
          "pot":"potassium",
          "hemo":"haemoglobin",
          "pcv":"packed_cell_volume",
          "wc":"white_blood_cell_count",
          "rc":"red_blood_cell_count",
          "htn":"hypertension",
          "dm":"diabetes_mellitus",
          "cad":"coronary_artery_disease",
          "appet":"appetite",
          "pe":"pedal_edema",
          "ane":"anemia"}

df.rename(columns=cols_names, inplace=True)
```

```python
numerical_features = []
categorical_features = []

for i in df.columns:
    if df[i].nunique()>7:
        numerical_features.append(i)
    else:
        categorical_features.append(i)
```

```python
for col in categorical_features:
    print(f'{col} has {df[col].unique()} categories.\n')
```

```python
df['diabetes_mellitus'] = df['diabetes_mellitus'].replace(to_replace = {'\tno':'no','\tyes':'yes',' yes':'yes'})
df['coronary_artery_disease'] = df['coronary_artery_disease'].replace(to_replace = '\tno', value='no')
df['classification'] = df['classification'].replace(to_replace = 'ckd\t', value = 'ckd')
```

```python
for col in categorical_features:
    print(f'{col} has {df[col].unique()} categories.\n')
```

```python
numerical_features
```

```python
for col in numerical_features:
    print(f'{col} has {df[col].unique()} values.\n')
```

```python
cols_names
```

```python
df[['hypertension','diabetes_mellitus','coronary_artery_disease','pedal_edema','anemia']] = df[['hypertension','diabetes_mellitus','coronary_artery_disease','pedal_edema','anemia']].replace(to_replace={'yes':1,'no':0})
df[['red_blood_cells','pus_cell']] = df[['red_blood_cells','pus_cell']].replace(to_replace={'abnormal':1,'normal':0})
df[['pus_cell_clumps','bacteria']] = df[['pus_cell_clumps','bacteria']].replace(to_replace={'present':1,'notpresent':0})
df[['appetite']] = df[['appetite']].replace(to_replace={'good':1,'poor':0})
df['classification'] = df['classification'].replace(to_replace={'ckd':1,'ckd\t':1,'notckd':0})
df.rename(columns={'classification':'class'},inplace=True)
df['coronary_artery_disease'] = df[['coronary_artery_disease']].replace(to_replace='\tno',value=0)
df['diabetes_mellitus'] = df['diabetes_mellitus'].replace(to_replace={'\tno':0,'\tyes':1,' yes':1})
df['packed_cell_volume'] = df['packed_cell_volume'].replace(to_replace={'\t?' : np.nan, '\t43' : '43'})
df['white_blood_cell_count'] = df['white_blood_cell_count'].replace(to_replace={'\t?' : np.nan, '\t6200' : '6200', '\t8400' : '8400'})
df['red_blood_cell_count'] = df['red_blood_cell_count'].replace(to_replace='\t?', value=np.nan)
```

```python
df.head()
```

```python
df.drop('id',axis=1,inplace=True)
```

```python
df['red_blood_cells'] = df['red_blood_cells'].fillna(df['red_blood_cells'].mode()[0])
df['red_blood_cell_count'] = df['red_blood_cell_count'].fillna(df['red_blood_cell_count'].mean)
df['white_blood_cell_count'] = df['white_blood_cell_count'].fillna(df['white_blood_cell_count'].mean)
df['potassium'] = df['potassium'].fillna(df['potassium'].mean)
df['sodium'] = df['sodium'].fillna(df['sodium'].mean)
df['packed_cell_volume'] = df['packed_cell_volume'].fillna(df['packed_cell_volume'].mean)
df['pus_cell'] = df['pus_cell'].fillna(df['pus_cell'].mode()[0])
df['haemoglobin'] = df['haemoglobin'].fillna(df['haemoglobin'].mean)
df['sugar'] = df['sugar'].fillna(df['sugar'].mode()[0])
df['specific_gravity'] = df['specific_gravity'].fillna(df['specific_gravity'].mode()[0])
df['albumin'] = df['albumin'].fillna(df['albumin'].mode()[0])
df['blood_glucose_random'] = df['blood_glucose_random'].fillna(round(df['blood_glucose_random'].mean(),0))
df['blood_urea'] = df['blood_urea'].fillna(round(df['blood_urea'].mean(),0))
df['serum_creatinine'] = df['serum_creatinine'].fillna(round(df['serum_creatinine'].mean(),2))
df['blood_pressure'] = df['blood_pressure'].fillna(round(df['blood_pressure'].mean(),0))
df['age'] = df['age'].fillna(round(df['age'].mean(),0))
```
