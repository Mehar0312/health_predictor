```python
import numpy as np
import pandas as pd
import pickle
```

```python
df = pd.read_csv('..\data\cancer.csv')
```

```python
df.isna().sum()
```

```python
pd.set_option('max_columns',33)
df.head()
```

```python
df.shape
```

```python
df.drop(df.columns[[0,-1]], axis=1, inplace=True)
```

```python
df.head(30)
```

```python
df.shape
```

Spliting feature and target data

```python
Xdata = df.drop(['diagnosis'], axis=1)
ydata = df['diagnosis']
```

Encoding target data

```python
yenc = np.asarray([1 if c == 'M' else 0 for c in ydata])
```

```python
df.columns
```

```python
cols = ['concave points_worst', 'perimeter_worst', 'concave points_mean', 'radius_worst', 'perimeter_mean', 'area_worst', 'radius_mean', 'area_mean', 'concavity_mean', 'concavity_worst']
```

```python
Xdata = df[cols]
Xdata.columns
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(Xdata, yenc, test_size=0.3, random_state=43, stratify=yenc)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
compare_data = pd.DataFrame(y_pred, columns=['y_pred'])
```

```python
y_test = np.array(y_test)
```

```python
compare_data['y_test'] = pd.DataFrame(y_test)
```

```python
compare_data
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
print('Accuracy : {}'.format(accuracy_score(y_test,y_pred)))
```

```python
confusion_matrix(y_test,y_pred)
```
